```python
import requests

# Define the URL to make the API request
url = "https://data.nasdaq.com/api/v3/datasets/FSE/AFX_X.json"
params = {
    "start_date": "2020-10-11",
    "end_date": "2020-10-11",  # Specify a single day for a sample
    "api_key": API_KEY  # Use your actual API key
}

response = requests.get(url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the JSON data
    sample_data = response.json()

    # Print a formatted version of the JSON data for better readability
    print(json.dumps(sample_data, indent=4))

else:
    print(f"Error in API request. Status code: {response.status_code}")
```

These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
#Task 1: Collect data from the Franfurt Stock Exchange for the ticker AFX_X for the whole year 2017
params = dict(start_date='2017-01-01', end_date='2017-12-31',api_key=API_KEY)
r = requests.get(url, params=params)
```

```python
# Task 2: Convert the returned JSON object into a Python dictionary
data = r.json()
print(type(r))
print(type(data))
```

```python
# Extract the column names and data
column_names = data['dataset']['column_names']
stock_data = data['dataset']['data']
print(column_names)
```

```python
# Create a dictionary with column names as keys and corresponding data as values
stock_dict = {column: [row[i] for row in stock_data] for i, column in enumerate(column_names)}
```

```python
# Task 3: Calculate what the highest and lowest opening prices were for the stock in this period.
opening_prices = [price for price in stock_dict['Open'] if price is not None]
highest_opening_price = max(opening_prices)
lowest_opening_price = min(opening_prices)
print("The highest opening price for AFX_X in 2017 was " + str(highest_opening_price) + ".")
print("The lowest opening price for AFX_X in 2017 was " +str(lowest_opening_price) + ".")
```

```python
# Alternativ way, Task #3: Highest and lowest opening prices for AFX_X over 2017
opens = []
for entry in data['dataset']['data']:
    open = entry[1]
    if open != None:
        opens.append(open)
print("The highest opening price for AFX_X in 2017 was " + str(max(opens)) + ".")
print("The lowest opening price for AFX_X in 2017 was " +str(min(opens)) + ".")
```

```python
# Task 4: What was the largest change in any one day (based on High and Low price)?
high_prices = stock_dict['High']
low_prices = stock_dict['Low']
daily_changes = [high - low for high, low in zip(high_prices, low_prices)]
largest_change_one_day = max(daily_changes)
print("The largest change in any one day in 2017 was " +str(largest_change_one_day))
```

```python
# Alternativ way, Task #4: Largest Change in any one day, based on High and Low Price
one_day_changes = []
for entry in data['dataset']['data']:
    high = entry[2]
    low = entry[3]
    one_day_changes.append(abs(high-low))
print("The largest change in any one day in 2017 was " +str(max(one_day_changes)))
```

```python
# Task 5: What was the largest change between any two days (based on Closing Price)?
closing_prices = stock_dict['Close']
changes_between_days = [abs(closing_prices[i] - closing_prices[i-1]) for i in range(1, len(closing_prices))]
largest_change_two_days = max(changes_between_days)
print("The largest change between any two days based on closing price, was " +str(largest_change_two_days) + ".")
```

```python
# Task 6: What was the average daily trading volume during this year?
trading_volumes = [volume for volume in stock_dict['Traded Volume'] if volume is not None]
average_daily_volume = sum(trading_volumes) / len(trading_volumes)
print("The average daily trading volume for AFX_X in 2017 was " + str(average_daily_volume))
```

```python
# Alternativ way, Task #6: Average Daily Trading Volume for the Year 2017
trading_volumes = []
for entry in data['dataset']['data']:
    if entry[6] != None:
        trading_volumes.append(entry[6])
avg_daily = sum(trading_volumes)/len(trading_volumes)
print("The average daily trading volume for AFX_X in 2017 was " + str(avg_daily) + ".")
```

```python
# Task 7 (Optional): What was the median trading volume during the year 2017.
sorted_volumes = sorted(trading_volumes)
n = len(sorted_volumes)
median_daily_volume = (sorted_volumes[n // 2] + sorted_volumes[(n - 1) // 2]) / 2
print("The median trading volume for AFX_X in 2017 was " + str(median_daily_volume) + ".")
```

```python
# Alternative way, Task #7: Median trading volume for the Year 2017
trading_volumes.sort()
length = len(trading_volumes)
# If the length is even, the median is the average of the two middle values
if len(trading_volumes) % 2 == 0:
    mid_high = length // 2
    mid_low = length // 2 - 1
    median = (trading_volumes[mid_high] + trading_volumes[mid_low]) / 2
else:
# If the length is odd, the median is the middle value
    mid = length // 2
    median = trading_volumes[mid]

print("The median daily trading volume for AFX_X in 2017 was " + str(median) + ".")
```
